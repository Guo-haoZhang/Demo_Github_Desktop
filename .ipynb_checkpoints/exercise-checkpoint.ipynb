```python
n=int(input('Enter the length of the list: '))
uList=[]
for i in range(0,n):
    ele=int(input('Enter the element of index {}: '.format(i)))
    uList.append(ele)
print(uList)
uList.sort(reverse=True)
print(uList)
rank = []
for x in uList:
    rank.append(uList.index(x)+1)
print(rank)
```

```python
# Returns the smallest and largest 
# elements of a sequence as a tuple
def min_max(t):
    return min(t), max(t)

seq = [12, 98, 23, 74, 3, 54]
print (min_max(seq))

string = 'She turned me into a newt!'
print (min_max(string))
myMin, myMax = min_max(string)
```
